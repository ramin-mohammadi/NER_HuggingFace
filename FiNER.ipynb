````python
import pandas as pd

a = pd.DataFrame([[0, 24], [3, 4], [3, 111], [1, 24], [2, 24]], columns=["bob", "some"])
print(a, "\n")
d = {0: "ss", 1: "b", 2: "r", 3: "d"}

a.bob = a.bob.map(d)
print(a)

label_list = list(a.bob.unique())
label_list.sort()
str_to_int = {l: i for i, l in enumerate(label_list)}
int_to_str = {i: l for (l, i) in str_to_int.items()}

print(label_list)
print(str_to_int)
print(int_to_str)
````
